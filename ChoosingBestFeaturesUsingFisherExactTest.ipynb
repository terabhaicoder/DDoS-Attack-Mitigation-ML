```python
import pandas as pd
file_path = 'your_new_file_path(without null).csv'

# Read CSV in chunks
chunk_size = 100000 

# Read CSV in chunks with low_memory=False
chunks = pd.read_csv(file_path, chunksize=chunk_size, low_memory=False)

# Initialize an empty DataFrame to store concatenated chunks
df = pd.DataFrame()
for chunk in chunks:
    df = pd.concat([df, chunk])
    
# Total number of rows in the DataFrame
total_rows = len(df)
print("Total number of rows:", total_rows)
```

```python
#FLOW FEATURES 

import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import RobustScaler
from sklearn.impute import SimpleImputer

# Define the flow features
flow_features = [
    'Flow Duration', 'Total Fwd Packets', 'Total Backward Packets',
    'Total Length of Fwd Packets', 'Total Length of Bwd Packets',
    'Flow Packets/s', 'Flow IAT Mean', 'Flow IAT Std',
    'Flow IAT Max', 'Flow IAT Min'
]

# Select flow features from the DataFrame
flow_df = df[flow_features]

# Replace infinity and too large values with NaN
flow_df.replace([np.inf, -np.inf], np.nan, inplace=True)


# Impute missing values
imputer = SimpleImputer(strategy='mean')
flow_df_imputed = imputer.fit_transform(flow_df)


# Scale the features using RobustScaler
scaler = RobustScaler()
X_scaled = scaler.fit_transform(flow_df_imputed)


# Define the target variable
y = df['Label']  # Replace 'Label' with the actual name of your target variable


# Number of features to select
k = 4

# Initialize SelectKBest with f_classif scoring function
selector = SelectKBest(score_func=f_classif, k=k)

# Fit selector to the data and transform it
X_selected = selector.fit_transform(X_scaled, y)

# Get the indices of the selected features
selected_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_features = [flow_features[i] for i in selected_indices]

# Print the selected features
print("Selected Features:", selected_features)

```
